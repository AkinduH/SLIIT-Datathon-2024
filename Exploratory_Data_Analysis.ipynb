```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import holoviews as hv
from holoviews import opts
```

```python
hv.extension('bokeh')  # Enable the Bokeh backend for interactive plots
```

```python
df_A = pd.read_csv("Datasets\df_A.csv")
df_B = pd.read_csv("Datasets\df_B.csv")
```

```python
unique_sensors = df_A["SENSOR_ID"].unique()
for sensor in unique_sensors:
    # Filter rows for the current sensor
    sensor_df = df_A[df_A["SENSOR_ID"] == sensor].copy()
    
    # Calculate differences
    sensor_df["DAILY_YIELD_diff"] = sensor_df["DAILY_YIELD"].diff()
    sensor_df["TOTAL_YIELD_diff"] = sensor_df["TOTAL_YIELD"].diff()
    
    # Apply filtering condition
    filtered_df = sensor_df[
        (sensor_df["DAILY_YIELD_diff"] != 0)
        & (sensor_df["TOTAL_YIELD_diff"] == 0)
        & (sensor_df["DAILY_YIELD"] != 0)
    ].copy()
    
    # Adjust DAILY_YIELD values
    filtered_df["DAILY_YIELD"] -= filtered_df["DAILY_YIELD_diff"]
    
    # Replace the original rows in df_A with filtered rows
    df_A.loc[filtered_df.index, :] = filtered_df.drop(
        ["DAILY_YIELD_diff", "TOTAL_YIELD_diff"], axis=1
    )
```

```python
unique_sensors = df_B["SENSOR_ID"].unique()
for sensor in unique_sensors:
    # Filter rows for the current sensor
    sensor_df = df_B[df_B["SENSOR_ID"] == sensor].copy()
    
    # Calculate differences
    sensor_df["DAILY_YIELD_diff"] = sensor_df["DAILY_YIELD"].diff()
    sensor_df["TOTAL_YIELD_diff"] = sensor_df["TOTAL_YIELD"].diff()
    
    # Apply filtering condition
    filtered_df = sensor_df[
        (sensor_df["DAILY_YIELD_diff"] != 0)
        & (sensor_df["TOTAL_YIELD_diff"] == 0)
        & (sensor_df["DAILY_YIELD"] != 0)
    ].copy()
    
    # Adjust DAILY_YIELD values
    filtered_df["DAILY_YIELD"] -= filtered_df["DAILY_YIELD_diff"]
    
    # Replace the original rows in df_A with filtered rows
    df_B.loc[filtered_df.index, :] = filtered_df.drop(
        ["DAILY_YIELD_diff", "TOTAL_YIELD_diff"], axis=1
    )
```

```python
def plot_data_on_time_series(df, features):  
    plots = []
    df['DATE_TIME'] = pd.to_datetime(df['DATE_TIME'])
    for feature in features:
        # Create a curve (line plot)
        curve = hv.Curve(
            df, 
            'DATE_TIME', 
            feature 
        ).opts(
            title=feature, xlabel='Date Time', ylabel=feature, width=800, height=300, 
            tools=['hover'], show_grid=True, xrotation=45
        )
        
        # Create scatter plot (dots for each data point)
        scatter = hv.Scatter(
            df, 
            'DATE_TIME', 
            feature 
        ).opts(
            size=5, color='red', tools=['hover'], show_grid=True
        )
        
        # Overlay the curve and scatter
        overlay = curve * scatter
        plots.append(overlay)
    return plots
```

```python
sensor_1_df = df_A[(df_A['SENSOR_ID'] == 'sensor 1')]

# Define features to plot
features = ['DAILY_YIELD', 'TOTAL_YIELD', 'DC_POWER', 'AC_POWER', 'AMBIENT_TEMPERATURE', 'MODULE_TEMPERATURE', 'IRRADIATION']

plots = plot_data_on_time_series(sensor_1_df, features)

# Combine plots vertically
layout = hv.Layout(plots).cols(1)

# Display Holoviews plot
layout
```
