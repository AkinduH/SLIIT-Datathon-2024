```python
def plot_data_on_time_series(df, features):  
    plots = []
    df['DATE_TIME'] = pd.to_datetime(df['DATE_TIME'])
    for feature in features:
        # Create a curve (line plot)
        curve = hv.Curve(
            df, 
            'DATE_TIME', 
            feature 
        ).opts(
            title=feature, xlabel='Date Time', ylabel=feature, width=800, height=300, 
            tools=['hover'], show_grid=True, xrotation=45
        )
        
        # Create scatter plot (dots for each data point)
        scatter = hv.Scatter(
            df, 
            'DATE_TIME', 
            feature 
        ).opts(
            size=5, color='red', tools=['hover'], show_grid=True
        )
        
        # Overlay the curve and scatter
        overlay = curve * scatter
        plots.append(overlay)
    return plots
```

```python
Plot_df = df_A[(df_A['SENSOR_ID'] == 'sensor 1')]

# Define features to plot
features = ['DAILY_YIELD', 'TOTAL_YIELD', 'DC_POWER', 'AC_POWER', 'AMBIENT_TEMPERATURE', 'MODULE_TEMPERATURE', 'IRRADIATION']

plots = plot_data_on_time_series(Plot_df, features)

# Combine plots vertically
layout = hv.Layout(plots).cols(1)

# Display Holoviews plot
layout
```

```python
from sklearn.feature_selection import mutual_info_regression

def make_mi_scores(X, y):
    X = X.copy()
    
    # Convert datetime columns to numeric
    for colname in X.select_dtypes(["datetime64"]):
        X[colname] = X[colname].view(int)  # Convert datetime to timestamp integers
    
    # Factorize categorical columns
    for colname in X.select_dtypes(["object", "category"]):
        X[colname], _ = X[colname].factorize()
    
    # All discrete features should now have integer dtypes
    discrete_features = [pd.api.types.is_integer_dtype(t) for t in X.dtypes]
    
    # Compute mutual information scores
    mi_scores = mutual_info_regression(X, y, discrete_features=discrete_features, random_state=0)
    mi_scores = pd.Series(mi_scores, name="MI Scores", index=X.columns)
    mi_scores = mi_scores.sort_values(ascending=False)
    
    return mi_scores

def plot_mi_scores(scores, title):
    scores = scores.sort_values(ascending=True)
    width = np.arange(len(scores))
    ticks = list(scores.index)
    plt.barh(width, scores)
    plt.yticks(width, ticks)
    plt.title(title)
    plt.show()
```

```python
Mi_df = A1_df.copy()
Mi_df.dropna(inplace=True)
X = Mi_df[['AMBIENT_TEMPERATURE', 'MODULE_TEMPERATURE', 'IRRADIATION', 'TIME', 'DC_POWER', 'AC_POWER']]
y = Mi_df['TOTAL_YIELD_diff']

mi_scores = make_mi_scores(X, y)

print(mi_scores)

plt.figure(dpi=100, figsize=(8, 5))
plot_mi_scores(mi_scores, "MI Scores for TOTAL_YIELD_diff")
```
